# Starship Titanic

In this post I document the code and techniques I used to solve the Kaggle competition problem Starship Titanic. 
The main Kaggle competition page can be found [here](https://www.kaggle.com/competitions/spaceship-titanic/overview)

Summary: To help rescue crews and retrieve the lost passengers, you are challenged to predict which passengers 
    were transported by the anomaly using records recovered from the spaceship’s damaged computer system.
    My code uses an ensamble of a Random Forest and a Neural Network. The data was cleaned by removing columns 
    that did not improve the model. Gaps in the dataset were iteratively filled using predictions from the 
    Random Forest.

Split Cabin into: deck, number and side.

```python
df_size = df['Cabin'].size

def prep_cabin(df):
  df_size = df['Cabin'].size
  deck = np.zeros(df_size, dtype=object)
  num = np.ma.zeros(df_size, dtype=int)
  side = np.zeros(df_size, dtype=object)

  for i in range(df_size):
    part = str(df['Cabin'][i]).split('/')
    if part[0] == 'nan':
      deck[i] = np.nan
      num[i] = np.ma.masked
      side[i] = np.nan
    else:
      deck[i] = part[0]
      num[i] = part[1]
      side[i] = part[2]
    
  df['CabinDeck'] = deck
  df['CabinNum'] = num
  df['CabinSide'] = side

  print(df['CabinDeck'].unique())
  deck_order = 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'
  df['CabinDeck'] = df['CabinDeck'].astype('category')
  df['CabinDeck'].cat.set_categories(deck_order, ordered=True, inplace=True)
  return df

df = prep_cabin(df)
```

Split Passenger ID to: Group and Number

```python
def prep_id(df):
  df_size = df['Cabin'].size
  gr_num = np.ma.zeros(df_size, dtype=int)
  p_num = np.ma.zeros(df_size, dtype=int)

  for i in range(df_size):
    part = str(df['PassengerId'][i]).split('_')
    if part[0] == 'nan':
      gr_num[i] = np.ma.masked
      p_num[i] = np.ma.masked
    else:
      gr_num[i] = int(part[0])
      p_num[i] = int(part[1])

  size_group = np.zeros((len(gr_num)))
  for i in range(len(gr_num)):
    size_group[i] = np.sum((gr_num == gr_num[i]).astype(int))

  df['IdIndividual'] = p_num
  df['IdGroup'] = gr_num
  df['GroupSize'] = size_group
  return df

df = prep_id(df)
```

Split Name is first and second

```python
print(df['Name'].unique())

def prep_name(df):
  df_size = df['Cabin'].size
  f_name = np.zeros(df_size, dtype=object)
  s_name = np.zeros(df_size, dtype=object)
  family = np.zeros(df_size, dtype=bool)
  gr_num = df['IdGroup']

  for i in range(df_size):
    part = str(df['Name'][i]).split()

    if part[0] == 'nan':
      f_name[i] = np.nan
      s_name[i] = np.nan
    else:
      f_name[i] = part[0]
      s_name[i] = part[1]

  df['FirstName'] = f_name
  df['SecondName'] = s_name

  for i in range(df_size):
    # find people in group and check if last name matches
    g_ind = gr_num[i] == gr_num
    if np.sum((s_name[i] == s_name[g_ind]).astype(int)) >= 2:
      family[i] = True

  df['FamilyGroup'] = family

  # Alphabetize names
  df['Name'] = df['Name'].astype(str)
  name_sort1 = set(sorted(df['Name']))
  df['Name'] = df['Name'].astype('category')
  df['Name'].cat.set_categories(name_sort1, ordered=True, inplace=True)

  df['FirstName'] = df['FirstName'].astype(str)
  name_sort2 = set(sorted(df['FirstName']))
  df['FirstName'] = df['FirstName'].astype('category')
  df['FirstName'].cat.set_categories(name_sort2, ordered=True, inplace=True)

  df['SecondName'] = df['SecondName'].astype(str)
  name_sort3 = set(sorted(df['SecondName']))
  df['SecondName'] = df['SecondName'].astype('category')
  df['SecondName'].cat.set_categories(name_sort3, ordered=True, inplace=True)
  return df

df = prep_name(df)
```

Alphabetize Names, HomePlanet and Destination

```python
print(df['HomePlanet'].unique())
print(df['Destination'].unique())

def prep_place(df):
  # Distance from Sun
  home_place = 'Earth', 'Mars', 'Europa'
  df['HomePlanet'] = df['HomePlanet'].astype('category')
  df['HomePlanet'].cat.set_categories(home_place, ordered=True, inplace=True)

  # Distance from Earth 40, 41, 80 light years
  place = 'TRAPPIST-1e', '55 Cancri e', 'PSO J318.5-22'
  df['Destination'] = df['Destination'].astype('category')
  df['Destination'].cat.set_categories(place, ordered=True, inplace=True)
  return df

df = prep_place(df)
```

## Make a Data Loader

```python
procs = [Categorify, FillMissing]

def random_split(size, per):
  target = int(size * per)
  rand_ind = np.random.randint(0, size, target)
  rand_ind = np.unique(rand_ind)
  while len(rand_ind) != target:
    rand_ind = np.append(rand_ind, np.random.randint(0, size, target - len(rand_ind)))
    rand_ind = np.unique(rand_ind)
  return rand_ind

rand_ind = random_split(df_size, 0.9)
rand_ind_t = rand_ind[:int(df_size * 0.1)]
rand_ind = rand_ind[int(df_size * 0.1):]
cond = np.zeros((df_size), dtype=bool)
cond_t = np.zeros((df_size), dtype=bool)
cond[rand_ind] = True
cond_t[rand_ind_t] = True
train_idx = np.where( cond)[0]
valid_idx = np.where(~(cond | cond_t))[0]
test_idx = np.where(cond_t)[0]
splits = (list(train_idx),list(valid_idx))

dep_var = 'Transported' # dependant variable
cont, cat = cont_cat_split(df, 1, dep_var=dep_var)
to = TabularPandas(df, procs, cat, cont, y_names=dep_var, splits=splits)
tt = TabularPandas(df.loc[test_idx], procs, cat, cont, y_names=dep_var)

len(to.train), len(to.valid), len(tt)
```

```python
def r_mse(pred,y): return round(math.sqrt(((pred-y)**2).mean()), 6)
def m_rmse(m, xs, y): return r_mse(m.predict(xs), y)
#def r_mspe(pred,y): return r_mse(pred, y) / (y.max() - y.min())
def r_mspe(pred,y): return r_mse(pred, y) / r_mse(y.mean(), y)
def m_rmspe(m, xs, y): return r_mspe(m.predict(xs), y)
```

## Clean the Data
The data can be cleaned by using the random forest to predict missing values.

```python
def add_missing_col(dfc, tt, xs2, procs, dep_var):
  xs1 = tt.xs
  c1 = list(xs1.columns)
  c2 = list(xs2.columns)
  changed = False
  for c in range(len(c2)):
    if c2[c] not in c1:
      if '_na' in c2[c]:
        # _na column wasn't made if there were no missing values
        #print(c2[c], c)
        dfc.insert(0, c2[c], 0)
        changed = True
      else:
        raise Exception('A Column in DataFrame is not present in other DataFrame') 
  if changed:
    cont, cat = cont_cat_split(dfc, 1, dep_var=dep_var)
    tt = TabularPandas(dfc, procs, cat, cont, y_names=dep_var)
  return tt

def fill_null(df, keep_thresh=0.6):
  procs = [Categorify, FillMissing]

  dep_var = 'Transported' # dependant variable
  dfc = df.drop(columns=[dep_var]).copy()

  # Loop over columns and split null values out then predict null values

  test_list = []
  for col in dfc.columns:
    test_list.append(dfc[col].isnull())

  cont_all, cat_all = cont_cat_split(dfc, 1)
  for reps in range(4):
    for col in dfc.columns:
      print(col)
      idx = np.nonzero(dfc.columns == col)[0][0]
      test_idx = np.array(test_list[idx]) # test_list is list of null values

      if np.sum(test_idx) == 0: # no nulls
        continue

      dfc_test = dfc.loc[test_idx]
      dfc_rest = dfc.loc[~test_idx]

      trig_cont = False
      for i in dfc_test.columns:
        if i == col:
          continue
        check_col = dfc_test[i].isnull()

        #print(i, check_col.all())
        if check_col.all():
          dfc_test = dfc_test.drop(columns=[i])
          dfc_rest = dfc_rest.drop(columns=[i])

      dfc_size = dfc_rest[col].size
      rand_ind = random_split(dfc_size, 0.9)
      cond = np.zeros((dfc_size), dtype=bool)
      cond[rand_ind] = True
      train_idx = np.where( cond)[0]
      valid_idx = np.where(~cond)[0]
      splits = (list(train_idx), list(valid_idx))

      dep_var = col # dependant variable
      cont, cat = cont_cat_split(dfc_rest, 1, dep_var=dep_var)
      to = TabularPandas(dfc_rest, procs, cat, cont, y_names=dep_var, splits=splits)
      cont, cat = cont_cat_split(dfc_test, 1, dep_var=dep_var)
      tt = TabularPandas(dfc_test, procs, cat, cont, y_names=dep_var)
      tt = add_missing_col(dfc_test, tt, to.train.xs, procs, dep_var)
      #print('train:', to.train.xs.columns)
      #print('test:', tt.xs.columns)

      xs, y = to.train.xs, to.train.y
      valid_xs, valid_y = to.valid.xs, to.valid.y

      m = rf(xs, y)
      print(m_rmspe(m, xs, y), m_rmspe(m, valid_xs, valid_y), tt.xs.shape)
      if m_rmspe(m, valid_xs, valid_y) >= keep_thresh: 
        # rmspe is relative to basic mean model
        # Less than 1 means better than mean guess.
        continue

      preds = np.stack([t.predict(tt.xs) for t in m.estimators_])
      if col in cont_all:
        preds = np.mean(preds, axis=0)
        if (col == 'CabinNum') | (col == 'IdGroup') | (col == 'IdIndividual'):
          preds = preds.astype(int)
        filled_var = np.array(dfc[col])
        filled_var[test_idx] = preds
      else:
        preds = stats.mode(preds, axis=0)[0][0]
        filled_var = np.array(dfc[col])
        #id_to_name = {i: name for i,name in enumerate(to.categorize.vocab)}
        #id_to_name = cat.decode(tt)[col]
        #id_to_name = to.categorize.vocab
        #id_to_name = to.categorize.decode(preds)
        print(filled_var.shape, test_idx.shape, np.sum(test_idx), len(preds.astype(int)), np.max(preds.astype(int)), len(to.categorize.vocab))
        try:
          filled_var[test_idx] = np.array(to.categorize.decode(np.round(preds)).replace('\'', '').replace(']', '').replace('[', '').replace(' ', '').split(','))
        except:
          continue

        c = np.zeros((dfc[col].size), dtype=bool)
        for i in range(dfc[col].size):
          c[i] = isinstance(dfc[col][i], bool)
        if any(c):
          filled_var[test_idx] = filled_var[test_idx].astype(bool) 

      #print(to.procs.classes[col]) # for independant variable
      #print(to.categorize.vocab) # for dependant variable

      dfc[col] = filled_var

  return dfc

dfc = fill_null(df)
```

```python
def prep_spend(dfc):
  # Total spend
  # This takes missing values for multiple columns so add it after filling missing values
  dfc['TotalSpend'] = dfc['RoomService'] + dfc['FoodCourt'] + dfc['ShoppingMall'] + dfc['Spa'] + dfc['VRDeck']
  return dfc

print(np.sum(dfc['Cabin'].isnull()), np.sum(dfc['CabinNum'].isnull()), np.sum(dfc['CabinDeck'].isnull()), np.sum(dfc['CabinSide'].isnull()))
def null_cabin(dfc):
  df_size = dfc['Cabin'].size
  list_null = [dfc['CabinDeck'].isnull(), dfc['CabinNum'].isnull(), dfc['CabinSide'].isnull(), dfc['Cabin'].isnull()]
  for i in range(df_size):
    if (list_null[0][i] | list_null[1][i] | list_null[2][i]) == False:
      if dfc.loc[i, 'Cabin'] != dfc.loc[i, 'CabinDeck'] + '/' + str(int(dfc.loc[i, 'CabinNum'])) + '/' + dfc.loc[i, 'CabinSide']:
        print(dfc.loc[i, 'Cabin'], dfc.loc[i, 'CabinDeck'], dfc.loc[i, 'CabinNum'], dfc.loc[i, 'CabinSide'])
        dfc.loc[i, 'Cabin'] = dfc.loc[i, 'CabinDeck'] + '/' + str(int(dfc.loc[i, 'CabinNum'])) + '/' + dfc.loc[i, 'CabinSide']
    elif list_null[3][i] == False:
      if dfc.loc[i, 'Cabin'].split('/')[0] != dfc.loc[i, 'CabinDeck']:
        dfc.loc[i, 'CabinDeck'] = dfc.loc[i, 'Cabin'].split('/')[0]
      if dfc.loc[i, 'Cabin'].split('/')[1] != str(dfc.loc[i, 'CabinNum']):
        dfc.loc[i, 'CabinNum'] = float(dfc.loc[i, 'Cabin'].split('/')[1])
      if dfc.loc[i, 'Cabin'].split('/')[1] != dfc.loc[i, 'CabinSide']:
        dfc.loc[i, 'CabinSide'] = dfc.loc[i, 'Cabin'].split('/')[2]
  return dfc

dfc = prep_spend(dfc)
dfc = null_cabin(dfc)
```

```python
dfc = dfc.join(df['Transported'])

rand_ind = random_split(df_size, 0.9)
rand_ind_t = rand_ind[:int(df_size * 0.1)]
rand_ind = rand_ind[int(df_size * 0.1):]
cond = np.zeros((df_size), dtype=bool)
cond_t = np.zeros((df_size), dtype=bool)
cond[rand_ind] = True
cond_t[rand_ind_t] = True
train_idx = np.where( cond)[0]
valid_idx = np.where(~(cond | cond_t))[0]
test_idx = np.where(cond_t)[0]
splits = (list(train_idx), list(valid_idx))

dep_var = 'Transported' # dependant variable
cont, cat = cont_cat_split(dfc, 1, dep_var=dep_var)
to = TabularPandas(dfc, procs, cat, cont, y_names=dep_var, splits=splits)
tt = TabularPandas(dfc.loc[test_idx], procs, cat, cont, y_names=dep_var)

xs, y = to.train.xs, to.train.y
valid_xs, valid_y = to.valid.xs, to.valid.y

m = rf(xs, y)
m_rmse(m, xs, y), m_rmse(m, valid_xs, valid_y)
```
